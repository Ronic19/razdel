# **JSON**

```python
import json
import pandas as pd
import pickle
```


1.1 Считайте файл contributors_sample.json. Воспользовавшись модулем json, преобразуйте содержимое файла в соответствующие объекты python. Выведите на экран информацию о первых 3 пользователях.

```python
with open('contributors_sample.json', 'r') as f:
  data = json.load(f)

data[:3]
```

1.2 Выведите уникальные почтовые домены, содержащиеся в почтовых адресах людей

```python
domains = set()
c = 0
for person in data:
  email = person['mail']
  domains.add(email[email.index('@') + 1:])
domains
```

1.3 Напишите функцию, которая по username ищет человека и выводит информацию о нем. Если пользователь с заданным username отсутствует, возбудите исключение ValueError

```python
def find_person(username):
  for p in data:
    if p['username'] == username:
      return p
  raise ValueError(f'нет пользователя с именем {username}')
```

```python
find_person('sheilaadams')
```

```python
# ValueError:
find_person('abc')
```

1.4 Посчитайте, сколько мужчин и женщин присутсвует в этом наборе данных.

```python
male = 0
female = 0
for p in data:
  if p['sex'] == 'M':
    male += 1
  else:
    female += 1
print(f'Всего мужчин: {male}\nВсего женщин: {female}')
```

1.5 Создайте pd.DataFrame contributors, имеющий столбцы id, username и sex.

```python
id = []
username = []
sex = []
for p in data:
  id.append(p['id'])
  username.append(p['username'])
  sex.append(p['sex'])

contributors = pd.DataFrame({'id': id, 'username': username, 'sex': sex})
contributors.head()
```

1.6 Загрузите данные из файла recipes_sample.csv (ЛР2) в таблицу recipes. Объедините recipes с таблицей contributors с сохранением строк в том случае, если информация о человеке отсутствует в JSON-файле. Для скольких человек информация отсутствует?

```python
recipes = pd.read_csv('recipes_sample.csv')
recipes = recipes.rename(columns={'id': 'recipe_id'})

df_merged = pd.merge(recipes, contributors, left_on='contributor_id', right_on='id', how='left')
df_merged.head()
```

```python
print(f'В JSON-файле отсутствует информация о {df_merged.isna().sum()["id"]} людях')
```

# **pickle**

2.1 На основе файла contributors_sample.json создайте словарь следующего вида:

{
    должность: [список username людей, занимавших эту должность]
}

```python
jobs_dict = {}
for p in data:
  for job in p['jobs']:
    username_list = jobs_dict.get(job, [])
    username_list.append(p['username'])
    jobs_dict[job] = username_list
jobs_dict
```

2.2 Сохраните результаты в файл job_people.pickle и в файл job_people.json с использованием форматов pickle и JSON соответственно. Сравните объемы получившихся файлов. При сохранении в JSON укажите аргумент indent.

```python
with open('job_people.pickle', 'wb') as fpkl:
  pickle.dump(jobs_dict, fpkl)
```

```python
with open('job_people.json', 'w') as fjson:
  json.dump(jobs_dict, fjson, indent=4)
```

```python
import os
json_size = os.path.getsize('job_people.json')
pickle_size = os.path.getsize('job_people.pickle')
print(f'Размер json-файла: {json_size}')
print(f'Размер pickle-файла: {pickle_size}')
```

```python
import xml
from bs4 import BeautifulSoup
```

3.1 По данным файла steps_sample.xml сформируйте словарь с шагами по каждому рецепту вида {id_рецепта: ["шаг1", "шаг2"]}. Сохраните этот словарь в файл steps_sample.json

```python
with open('steps_sample.xml') as fr:
  steps = BeautifulSoup(fr, 'xml')

steps
```

```python
id_steps = {}
c = 0
for r in steps.find_all('recipe'):
  list_steps = [k.next for k in r.steps.find_all('step')]
  id_steps[r.id.next] = list_steps

id_steps
```

```python
with open('steps_sample.json', 'w') as fjson:
  json.dump(id_steps, fjson)
```

3.2 По данным файла steps_sample.xml сформируйте словарь следующего вида: кол-во_шагов_в_рецепте: [список_id_рецептов]

```python
n_steps = {}
for k, v in id_steps.items():
  id_list = n_steps.get(len(v), [])
  id_list.append(k)
  n_steps[len(v)] = id_list
print(n_steps)
```

3.3 Получите список рецептов, в этапах выполнения которых есть информация о времени (часы или минуты). Для отбора подходящих рецептов обратите внимание на атрибуты соответствующих тэгов.

```python
res = []
for r in steps.find_all('recipe'):
  for ss in r.find_all('steps'):
    for st in ss.find_all('step'):
      if st.get('has_minutes') or st.get('has_hours'):
        res.append(r.id.next)
res
```

3.4 Загрузите данные из файла recipes_sample.csv (ЛР2) в таблицу recipes. Для строк, которые содержат пропуски в столбце n_steps, заполните этот столбец на основе файла steps_sample.xml. Строки, в которых столбец n_steps заполнен, оставьте без изменений.

```python
recipes = pd.read_csv('recipes_sample.csv')
count_steps = {k: len(v) for k, v in id_steps.items()} # id_steps - словарь из задания 3.1
recipes.head()
```

```python
df_count_steps = pd.DataFrame({'id': [int(k) for k in count_steps.keys()], 'n_steps': [v for v in count_steps.values()]})
na_recipes = recipes[recipes['n_steps'].isna()].drop('n_steps', axis=1)
df_merged = pd.merge(na_recipes, df_count_steps, left_on='id', right_on='id')
bool_list = [i == False for i in recipes['n_steps'].isna()]
new_df = df_merged.append(recipes[bool_list])
new_df
```
