5.1 При помощи объединения таблиц, создайте DataFrame, состоящий из четырех столбцов: id, name, user_id, rating. Рецепты, на которые не оставлен ни один отзыв, должны отсутствовать в полученной таблице. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и попытавшись найти строку, соответствующую этому рецепту, в полученном DataFrame.

```python
recipes_1 = recipes[['id', 'name']]
reviews_1 = reviews[['recipe_id', 'user_id', 'rating']]
df_merged = pd.merge(recipes_1, reviews_1, left_on='id', right_on='recipe_id')
df_merged.drop('recipe_id', axis=1, inplace=True)
df_merged
```

```python
without_reviews = [i for i in recipes['id'].values if not i in reviews['recipe_id'].values] # список рецептов, на которые нет отзывов
# пытаемся найти в df_merged рецепт без отзывов:
df_merged[df_merged['id'] == without_reviews[4]]
```

5.2 При помощи объединения таблиц и группировок, создайте DataFrame, состоящий из трех столбцов: recipe_id, name, review_count, где столбец review_count содержит кол-во отзывов, оставленных на рецепт recipe_id. У рецептов, на которые не оставлен ни один отзыв, в столбце review_count должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и найдя строку, соответствующую этому рецепту, в полученном DataFrame.

```python
recipe_2 = recipes[['id', 'name']].copy()
reviews_2 = reviews[['recipe_id']].copy()
reviews_2['review_count'] = range(reviews_2.shape[0])
df_merged_1 = pd.merge(pd.DataFrame(reviews_2.groupby('recipe_id')['review_count'].count()), recipe_2, left_on='recipe_id', right_on='id', how='outer')
df_merged_1.fillna(0.0, inplace=True)
df_merged_1
```

```python
# выводим рецепт, у которого нет отзывов
df_merged_1[df_merged_1['id'] == without_reviews[5]]
```

5.3. Выясните, рецепты, добавленные в каком году, имеют наименьший средний рейтинг?

```python
recipes_3 = recipes[['id', 'submitted']]
reviews_3 = reviews[['recipe_id', 'rating']]
df_merged_2 = pd.merge(recipes_3, reviews_3, left_on='id', right_on='recipe_id', how='outer')
df_merged_2
```

```python
year_rating = df_merged_2.groupby(df_merged_2['submitted'].dt.year)['rating'].mean()
print(f'Минимальный средний рейтинг ({year_rating.min()}) у рецептов, добавленных в {year_rating[year_rating == year_rating.min()].index[0]} году')
```
